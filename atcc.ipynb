````python
import cv2
from ultralytics import YOLO
import time
from google.colab.patches import cv2_imshow

# Load the YOLO model
def load_model(model_path="yolov8n.pt"):  # Use a pre-trained YOLOv8 model
    return YOLO(model_path)

# Initialize video captures
def initialize_video_captures(video_paths):
    return [cv2.VideoCapture(path) for path in video_paths]

# Detect vehicles in a frame
def detect_vehicles(frame, model, vehicle_classes):
    results = model(frame)
    detections = results[0].boxes.cls.numpy()
    vehicle_count = sum(cls in vehicle_classes for cls in detections)
    return vehicle_count

# Update traffic lights based on traffic counts
def update_traffic_lights(traffic_counts):
    current_green_road = traffic_counts.index(max(traffic_counts))
    return current_green_road

# Process video streams and perform detection
def process_videos(video_captures, model, vehicle_classes, green_duration, yellow_duration):
    traffic_counts = [0] * len(video_captures)
    frame_counter = 0
    resize_width, resize_height = 648, 360

    # Video writer setup (Optional: Save combined output video)
    out = cv2.VideoWriter('output.avi', cv2.VideoWriter_fourcc(*'XVID'), 10, (resize_width * 2, resize_height * 2))

    while all(cap.isOpened() for cap in video_captures):
        frames = []
        overlay_texts = []

        for i, cap in enumerate(video_captures):
            ret, frame = cap.read()
            if not ret:
                print(f"Video {i+1} ended.")
                cap.release()
                continue

            # Resize the frame
            frame = cv2.resize(frame, (resize_width, resize_height))

            # Detect vehicles
            traffic_counts[i] = detect_vehicles(frame, model, vehicle_classes)

            # Update traffic light status
            current_green_road = update_traffic_lights(traffic_counts)
            status = "GREEN" if i == current_green_road else "RED"
            overlay_text = f"Road {i+1} | Vehicles: {traffic_counts[i]} | Status: {status}"
            overlay_texts.append(overlay_text)

            # Overlay the text on the frame
            cv2.putText(frame, overlay_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            frames.append(frame)

        # Ensure all frames are processed
        if len(frames) < len(video_captures):
            break

        # Combine frames into a grid (2x2)
        top_row = cv2.hconcat([frames[0], frames[1]])
        bottom_row = cv2.hconcat([frames[2], frames[3]])
        combined_frame = cv2.vconcat([top_row, bottom_row])

        # Display the combined frame
        cv2_imshow(combined_frame)

        # Save the combined frame to the output video
        out.write(combined_frame)

        # Exit on key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release all resources
    for cap in video_captures:
        cap.release()
    out.release()
    cv2.destroyAllWindows()

# Main code execution
if __name__ == "__main__":
    # Define paths to your video files
    video_paths = [
        "1952-152220070_tiny.mp4",
        "27260-362770008_tiny.mp4",
        "45527-443134238_tiny.mp4",
        "BB_7bee62f7-aa19-4a88-a849-041f232de78a.mp4"
    ]

    # Vehicle classes (YOLO class IDs for cars, motorcycles, buses, etc.)
    vehicle_classes = [2, 3, 5, 7]  # Cars, Motorcycles, Buses, Trucks

    # Light durations
    yellow_light_duration = 3
    green_light_duration = 10

    # Load YOLO model
    model = load_model()

    # Initialize video captures
    video_captures = initialize_video_captures(video_paths)

    # Process the videos
    process_videos(video_captures, model, vehicle_classes, green_light_duration, yellow_light_duration)
````
